### Tratamento dos dados --> Corrigindo alguns problemas que a nossa base da dados possui

```python
### 1° COLUNAS
# temos uma coluna (Data_Base) na nossa base de dados que é inútil (pois ela esta cheia de valores nulos)
'''
vendas_df.drop(columns=["coluna1", "coluna2", ..., "coluna_n"])
'''
df_analise = vendas_df.drop(columns=["Data_Base"])
# display(df_analise)

### 2° DADOS NULOS

# df_analise = df_analise.dropna()  # exlcui todas as linhas que possuem PELO MENOS 1 valor vazio (maneira mais simples de tratar valores vazio)

# vimos que a unica que coluna (tirando a Data_Base que ja excluimos) que tem alguns valores vazio é a "Loja".
# vamos então, ao inves de apagar as linhas aonde essa coluna é nula, preencher algum valor jsutamente nessa entrada que é nula
df_analise["Loja"] = df_analise["Loja"].fillna("Online") # vamos assumir que quando a loja estava vazio era na vrdd uma compra em uma loja Online
# display(df_analise)



### 3° TIPOS DE DADOS
# a coluna "Data" esta sendo reconhecida como texto, e não como data
df_analise["Data"] = pd.to_datetime(df_analise["Data"], format="%Y-%m-%d")
# display(df_analise.info())

### 4° PADRONIZAÇÃO
df_analise['Loja'] = df_analise['Loja'].str.strip() # remove espaços no inicio e no final do texto
df_analise['Loja'] = df_analise['Loja'].str.title() # padroniza o texto com a primeira letra de cada palavra maiuscula e as demais minusculas



### 4° DUPLICATAS
# o parametro subset() dentro do drop_duplicates() é para voce selecionar quais colunas serão consideradas na analise de valores iguais
# exemplo: suponha que voce tenha duas linhas onde o unico valor que se repete é da coluna "ID_Pedido". com o drop_duplicates(), normal
# ou seja, sem esse parametro do subset, essas duas linhas não vao ser consideradas duplicata. logo, não vao ser removidas
# ai dentro do subset() vc passa quantas colunas vc quer considerar
# o parametro keep é para voce escolher qual linha o pandas ira manter dentre as linhas consideradas iguais
df_analise = df_analise.drop_duplicates(subset=["ID_Pedido"], keep="first") # remove as linhas que são iguais (os mesmo valores em todas as colunas)

display(df_analise)
# display(df_analise.info())
```
