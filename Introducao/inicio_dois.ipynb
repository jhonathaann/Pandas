### Tratamento dos dados --> Corrigindo alguns problemas que a nossa base da dados possui

```python
### 1° COLUNAS
# temos uma coluna (Data_Base) na nossa base de dados que é inútil (pois ela esta cheia de valores nulos)
'''
vendas_df.drop(columns=["coluna1", "coluna2", ..., "coluna_n"])
'''
df_analise = vendas_df.drop(columns=["Data_Base"])
# display(df_analise)

### 2° DADOS NULOS

# df_analise = df_analise.dropna()  # exlcui todas as linhas que possuem PELO MENOS 1 valor vazio (maneira mais simples de tratar valores vazio)

# vimos que a unica que coluna (tirando a Data_Base que ja excluimos) que tem alguns valores vazio é a "Loja".
# vamos então, ao inves de apagar as linhas aonde essa coluna é nula, preencher algum valor jsutamente nessa entrada que é nula
df_analise["Loja"] = df_analise["Loja"].fillna("Online") # vamos assumir que quando a loja estava vazio era na vrdd uma compra em uma loja Online
# display(df_analise)



### 3° TIPOS DE DADOS
# a coluna "Data" esta sendo reconhecida como texto, e não como data
df_analise["Data"] = pd.to_datetime(df_analise["Data"], format="%Y-%m-%d")
# display(df_analise.info())

### 4° PADRONIZAÇÃO
df_analise['Loja'] = df_analise['Loja'].str.strip() # remove espaços no inicio e no final do texto
df_analise['Loja'] = df_analise['Loja'].str.title() # padroniza o texto com a primeira letra de cada palavra maiuscula e as demais minusculas

gerentes_df['Loja'] = gerentes_df['Loja'].str.strip() # remove espaços no inicio e no final do texto
gerentes_df['Loja'] = gerentes_df['Loja'].str.title() # padroniza o texto com a primeira letra de cada palavra maiuscula e as demais minusculas



### 4° DUPLICATAS
# o parametro subset() dentro do drop_duplicates() é para voce selecionar quais colunas serão consideradas na analise de valores iguais
# exemplo: suponha que voce tenha duas linhas onde o unico valor que se repete é da coluna "ID_Pedido". com o drop_duplicates(), normal
# ou seja, sem esse parametro do subset, essas duas linhas não vao ser consideradas duplicata. logo, não vao ser removidas
# ai dentro do subset() vc passa quantas colunas vc quer considerar
# o parametro keep é para voce escolher qual linha o pandas ira manter dentre as linhas consideradas iguais
df_analise = df_analise.drop_duplicates(subset=["ID_Pedido"], keep="first") # remove as linhas que são iguais (os mesmo valores em todas as colunas)

display(df_analise)
# display(df_analise.info())
```

### Criando novas colunas

```python
### FATURAMENTO
df_analise["Faturamento"] = df_analise["Preco_Unitario"] * df_analise["Qtd"]


### FORMA DE VENDA --> queremos que essa coluna informe quando a venda é online ou não
## np.where(condicao, valor_verdadeira, valor_falso)
df_analise["Forma_de_Venda"] = np.where(df_analise["Loja"] == "Online", "Online", "Presencial")
# display(df_analise)


### REGIAO
# Como eu descubro quantas regiões existem? para isso, eu preciso saber quais sãos os valores distintos que essa coluna possui

# criando um dicionario para realizar o mapeamento entre cada cidade com a sua respectiva regiao
dic_regioes = {
    'São Paulo' : "Sudeste", 
    'Belo Horizonte' : "Sudeste", 
    'Online' : "Online", 
    'Rio De Janeiro' : "Sudeste",
    'Salvador' : "Nordeste", 
    'Recife' : "Nordeste", 
    'Curitiba' : "Sul", 
    'Porto Alegre' : "Sul"
} 
# display(df_analise["Loja"].unique())

# 1° forma de fazer isso: usando o map() com o auxilio de uma funcao

# def definir_regioes(loja):
#   dic_regioes = {
#     'São Paulo' : "Sudeste", 
#     'Belo Horizonte' : "Sudeste", 
#     'Online' : "Online", 
#     'Rio De Janeiro' : "Sudeste",
#     'Salvador' : "Nordeste", 
#     'Recife' : "Nordeste", 
#     'Curitiba' : "Sul", 
#     'Porto Alegre' : "Sul"
#   }

#   return dic_regioes[loja]

# df_analise["Região"] = df_analise["Loja"].map(definir_regioes)

# 2° forma é uma maneira mais simples: passar o dicionario dentro do map
df_analise["Região"] = df_analise["Loja"].map(dic_regioes)


display(df_analise)

### Como eu descubro se funcionou? no caso, vamos ver se existem algum valor vazio no nosso df
display(df_analise.isna().sum())
```

### Analises por agrupamento

```python
### Pegando a loja que mais vendeu

# groupy() --> ele agrupa a coluna que voce passa dentro do ()
# por exemplo, se eu agrupar por loja, cada loja so vai aparecer uma vez

analise_lojas = df_analise[["Loja", "Faturamento"]].groupby("Loja").sum() # OBS: por padrão a "Loja" vai ser usada como indice no
# novo df montado. o parametro as_index=False do groupby tira isso (mas vamos fazer isso de uma outra forma, como segue abaixo)
analise_lojas = analise_lojas.sort_values(by="Faturamento", ascending=False)
analise_lojas = analise_lojas.reset_index()

# formatando agora a coluna faturamento
analise_lojas["Faturamento"] = analise_lojas["Faturamento"].map("R${:,.2f}".format)

# a linha acima faz: ela agrupa a coluna "Loja" somando os valores da coluna "Faturamento", ou seja, como cada loja so vai
# aparecer uma vez, o que vai acontecer com a coluna "Faturamento" das linhas em que a coluna "Loja" é igual? esses valores vao ser somados!!
# display(analise_lojas)


### RANKING DOS PRODUTOS QUE MAIS VENDERAM ONLINE
df_produtos_online = df_analise[df_analise["Loja"]=="Online"]
analise_produtos_online = df_produtos_online[["Produto", "Qtd"]].groupby("Produto").sum()
analise_produtos_online = analise_produtos_online.sort_values(by="Qtd", ascending=False)
analise_produtos_online = analise_produtos_online.rename(columns={"Qtd" : "Quantidade"})
display(analise_produtos_online)
```

### Analise por mais de um agrupamento

```python
### analise por loja e por produto

## 1° quais produtos venderam mais em cada uma das lojas
df_analise_produtos_loja = df_analise[["Loja", "Produto","Qtd"]].groupby(["Loja", "Produto"]).sum() # aqui ele vai agrupar
# primeiro por cada uma das lojas E depois por cada produto (somando a quantidade de produto)
display(df_analise_produtos_loja)

## 2° quais lojas mais venderam os produtos
df_analise_lojas_em_produto = df_analise[["Loja", "Produto","Qtd"]].groupby(["Produto", "Loja"]).sum()
display(df_analise_lojas_em_produto)
```
